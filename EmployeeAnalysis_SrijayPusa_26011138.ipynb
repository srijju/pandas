TASK1 : Create Database

```python
import sqlite3
import random
import pandas as pd

conn = sqlite3.connect('employees.db')
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    emp_id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary REAL,
    years_experience INTEGER,
    performance_score REAL
)
""")

departments = ["Engineering", "Sales", "Marketing", "HR", "Finance"]

employees_data = []

for i in range(1, 41):
    name = f"Employee_{i}"
    department = random.choice(departments)
    salary = random.randint(50000, 150000)
    years_experience = random.randint(1, 15)
    performance_score = round(random.uniform(1.0, 5.0), 2)

    employees_data.append((i, name, department, salary, years_experience, performance_score))

cursor.executemany("""
INSERT INTO employees VALUES (?, ?, ?, ?, ?, ?)
""", employees_data)

conn.commit()
```

TASK2: SQL Queries

```python

query1 = """
SELECT name, department, salary, performance_score
FROM employees
WHERE performance_score >= 4.0
AND years_experience >= 3
ORDER BY performance_score DESC
LIMIT 15
"""

df_q1 = pd.read_sql_query(query1, conn)

df_q1
```

```python
query2 = """
SELECT *
FROM employees
WHERE salary BETWEEN 70000 AND 110000
AND department IN ('Engineering','Sales')
ORDER BY department ASC, salary DESC
"""

df_q2 = pd.read_sql_query(query2, conn)

df_q2
```

```python
query3 = """
SELECT department,
COUNT(*) AS employee_count,
AVG(salary) AS avg_salary
FROM employees
GROUP BY department
ORDER BY avg_salary DESC
"""

df_q3 = pd.read_sql_query(query3, conn)

df_q3
```

TASK3: Pandas Analysis

```python
df = pd.read_sql_query("SELECT * FROM employees", conn)

avg_performance = df.groupby("department")["performance_score"].mean()

avg_performance
```

```python
df_query1_pandas = (
    df.loc[(df["performance_score"] >= 4.0) & (df["years_experience"] >= 3),
           ["name","department","salary","performance_score"]]
    .sort_values(by="performance_score", ascending=False)
    .head(15)
)

df_query1_pandas
```

SQL and Pandas are both powerful tools for data analysis but are used in different contexts. SQL is a query language designed for interacting with relational databases, while Pandas is a Python library used for in-memory data manipulation and analysis.

One major difference between SQL and Pandas is their syntax. SQL uses expressions like = for comparison, AND for combining conditions, and ORDER BY for sorting results. In contrast, Pandas uses Python syntax such as == for equality checks, & for combining conditions, and the .sort_values() method for sorting. Filtering in Pandas is typically done using .loc[] or boolean indexing.

SQL is best used when working directly with databases because it allows efficient querying without loading all data into memory. It is ideal for tasks such as aggregations, filtering large datasets, and retrieving specific subsets of data.

Pandas, on the other hand, is better suited for advanced data manipulation, statistical analysis, and data visualization within Python environments such as Jupyter or Google Colab. It provides flexible tools like groupby(), merge(), and apply() for complex operations.

For datasets that are too large to fit into memory, SQL is generally the better choice because database systems can process large data efficiently using optimized query engines and disk storage rather than relying entirely on RAM.
